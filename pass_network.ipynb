```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.colors import to_rgba
from mplsoccer import Pitch, FontManager, Sbopen
from statsbombpy import sb
```

```python
sb.competitions()
```

```python
sb.matches(competition_id=43, season_id=106)
```

```python
events = sb.events(match_id=3857258)
```

```python
events.head()
```

```python
events.columns
```

```python
events = events[['id', 'minute', 'second', 'team', 'type', 'pass_outcome', 'player', 'player_id', 'location', 'pass_end_location']]
subs = events[events['type']=='Substitution']
subs = subs['minute']
firstSub = subs.min()
events = events[events['team']=='Brazil']
events = events[events['pass_outcome']!='Incomplete']
events = events[events['pass_outcome']!='Injury Clearance']
events = events[events['type']=='Pass']
```

```python
events.head(1000)
```

```python
lineup = sb.lineups(match_id=3857258)["Brazil"]
lineup.head()
```

```python
events = pd.merge(events, lineup, left_on='player', right_on='player_name')
```

```python
events.head()
```

```python
events = events.drop(columns=['player', 'cards', 'positions', 'country', 'player_id_x'])
```

```python
events = events.sort_values(by=['minute', 'second'])
```

```python
events['passer'] = events['jersey_number']
events['recipient'] = events['jersey_number'].shift(-1)

pas = pd.to_numeric(events['passer'], downcast='integer')
rec = pd.to_numeric(events['recipient'], downcast='integer')

events['passer'] = pas
events['recipient'] = rec
```

```python
events.head()
```

```python
firstSub
```

```python
events = events[events['minute']<firstSub]
```

```python
events[['x', 'y']] = pd.DataFrame(events['location'].tolist(), index=events.index)
events[['end_x', 'end_y']] = pd.DataFrame(events['pass_end_location'].tolist(), index=events.index)
```

```python
events
```

```python
# find avg locations
avg_loc = events.groupby('passer').agg({'x':['mean'], 'y':['mean', 'count']})
avg_loc.columns = ['x', 'y', 'count']
avg_loc
```

```python
# find num of passes between each player
pass_between = events.groupby(['passer', 'recipient']).id.count().reset_index()
pass_between.rename({'id':'pass_count'}, axis='columns', inplace=True)
```

```python
pass_between = pass_between.merge(avg_loc, left_on='passer', right_index = True)
pass_between = pass_between.merge(avg_loc, left_on='recipient', right_index = True, suffixes=['', '_end'])
```

```python
pass_between
```

```python
pass_between = pass_between[pass_between['pass_count']>2]
```

```python
pass_between
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor("#22312b")

MIN_TRANSPARENCY = 0.3
color = np.array(to_rgba('white'))
color = np.tile(color, (len(pass_between), 1))
c_transparency = pass_between.pass_count / pass_between.pass_count.max()
c_transparency = (c_transparency * (1 - MIN_TRANSPARENCY)) + MIN_TRANSPARENCY
color[:, 3] = c_transparency

MAX_LINE_WIDTH = 18
MAX_MARKER_SIZE = 3000
pass_between['width'] = (pass_between.pass_count / pass_between.pass_count.max() * MAX_LINE_WIDTH)
pass_between['marker_size'] = (pass_between['count'] / pass_between['count'].max() * MAX_MARKER_SIZE)

pass_lines = pitch.lines(pass_between.x, pass_between.y,
                         pass_between.x_end, pass_between.y_end, color=color, zorder=1, ax=ax)
pass_nodes = pitch.scatter(pass_between.x, pass_between.y,
                           s=pass_between.marker_size,
                           color='red', edgecolors='black', linewidth=1, alpha=1, ax=ax)

for index, row in pass_between.iterrows():
    pitch.annotate(row.passer, xy=(row.x, row.y), c='white', va='center',
                   ha='center', size=16, weight='bold', ax=ax)
```

```python
fig, axs = pitch.grid(figheight=10, title_height=0.08, endnote_space=0,
                      # Turn off the endnote/title axis. I usually do this after
                      # I am happy with the chart layout and text placement
                      axis=False,
                      title_space=0, grid_height=0.82, endnote_height=0.05)
fig.set_facecolor("#22312b")

pass_lines = pitch.lines(pass_between.x, pass_between.y,
                         pass_between.x_end, pass_between.y_end, lw=pass_between.width,
                         color=color, zorder=1, ax=axs['pitch'])
pass_nodes = pitch.scatter(pass_between.x, pass_between.y,
                           s=pass_between.marker_size,
                           color='red', edgecolors='black', linewidth=1, alpha=1, ax=axs['pitch'])
for index, row in pass_between.iterrows():
    pitch.annotate(row.passer.astype(int), xy=(row.x, row.y), c='white', va='center',
                   ha='center', size=16, weight='bold', ax=axs['pitch'])
    
TEAM = 'Brazil'
OPPONENT = 'Serbia'
FORMATION = 4231

# Load a custom font.
URL = 'https://raw.githubusercontent.com/google/fonts/main/apache/roboto/Roboto%5Bwdth,wght%5D.ttf'
robotto_regular = FontManager(URL)

# endnote /title
axs['endnote'].text(1, 0.5, '@bradyballFC', color='#c7d5cc',
                    va='center', ha='right', fontsize=15,
                    fontproperties=robotto_regular.prop)
TITLE_TEXT = f'{TEAM}, {FORMATION} formation'
axs['title'].text(0.5, 0.7, TITLE_TEXT, color='#c7d5cc',
                  va='center', ha='center', fontproperties=robotto_regular.prop, fontsize=30)

SECONDARY_TITLE = f'versus {OPPONENT}, 2022 World Cup - Group G'
axs['title'].text(0.5, 0.25, SECONDARY_TITLE, color='#c7d5cc',
                  va='center', ha='center', fontproperties=robotto_regular.prop, fontsize=18)

plt.show()
```
